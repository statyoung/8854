```python
scaled_df = df.copy()
```

```python
#scaled_df.hist(figsize=(20, 20), bins=50, xlabelsize=8, ylabelsize=8, log=True);
#scaled_df.describe()
```

```python
scaled_df['indegree'] = np.log(scaled_df['indegree']).replace(0, 0.1).replace(-np.inf, 0)
scaled_df['outdegree']= np.log(scaled_df['outdegree']).replace(0, 0.1).replace(-np.inf, 0)
# # #df['in_btc'] = np.log(df['in_btc']).replace(-np.inf, 0)
# # #df['out_btc']= np.log(df['out_btc']).replace(-np.inf, 0)
scaled_df['total_btc']= np.log(scaled_df['total_btc']).replace(0, 0.1).replace(-np.inf, 0)
scaled_df['mean_in_btc']= np.log(scaled_df['mean_in_btc']).replace(0, 0.1).replace(-np.inf, 0)
scaled_df['mean_out_btc']= np.log(scaled_df['mean_out_btc']).replace(0, 0.1).replace(-np.inf, 0)

scaled_df.describe()
```

```python
norm_scaled_df = scaled_df.copy()
norm_scaled_df['indegree'] = preprocessing.Normalizer(norm='l1').fit_transform(norm_scaled_df['indegree'].values.reshape(-1, 1))
norm_scaled_df['outdegree'] = preprocessing.Normalizer(norm='l1').fit_transform(norm_scaled_df['outdegree'].values.reshape(-1, 1))
#scaled_df['in_btc'] = preprocessing.Normalizer().fit_transform(df['in_btc'].values.reshape(-1, 1))
#scaled_df['out_btc'] = preprocessing.Normalizer().fit_transform(df['out_btc'].values.reshape(-1, 1))
norm_scaled_df['total_btc'] = preprocessing.Normalizer(norm='l1').fit_transform(norm_scaled_df['total_btc'].values.reshape(-1, 1))
norm_scaled_df['mean_in_btc'] = preprocessing.Normalizer(norm='l1').fit_transform(norm_scaled_df['mean_in_btc'].values.reshape(-1, 1))
norm_scaled_df['mean_out_btc'] = preprocessing.Normalizer(norm='l1').fit_transform(norm_scaled_df['mean_out_btc'].values.reshape(-1, 1))
norm_scaled_df.describe()
```

```python
mx_scaled_df = scaled_df.copy()
mx_scaled_df['indegree'] = preprocessing.MinMaxScaler(feature_range=(-1, 1)).fit_transform(mx_scaled_df['indegree'].values.reshape(-1, 1))
mx_scaled_df['outdegree'] = preprocessing.MinMaxScaler(feature_range=(-1, 1)).fit_transform(mx_scaled_df['outdegree'].values.reshape(-1, 1))
#scaled_df['in_btc'] = StandardScaler().fit_transform(df['in_btc'].values.reshape(-1, 1))
#scaled_df['out_btc'] = StandardScaler().fit_transform(df['out_btc'].values.reshape(-1, 1))
mx_scaled_df['total_btc'] = preprocessing.MinMaxScaler(feature_range=(-1, 1)).fit_transform(mx_scaled_df['total_btc'].values.reshape(-1, 1))
mx_scaled_df['mean_in_btc'] = preprocessing.MinMaxScaler(feature_range=(-1, 1)).fit_transform(mx_scaled_df['mean_in_btc'].values.reshape(-1, 1))
mx_scaled_df['mean_out_btc'] = preprocessing.MinMaxScaler(feature_range=(-1, 1)).fit_transform(mx_scaled_df['mean_out_btc'].values.reshape(-1, 1))
mx_scaled_df.describe()
```

```python
rb_scaled_df = scaled_df.copy()
rb_scaled_df['indegree'] = preprocessing.RobustScaler(quantile_range=(25.0, 75.0), with_centering=True,
       with_scaling=True).fit_transform(rb_scaled_df['indegree'].values.reshape(-1, 1))

rb_scaled_df['outdegree'] = preprocessing.RobustScaler(quantile_range=(25.0, 75.0), with_centering=True,
       with_scaling=True).fit_transform(rb_scaled_df['outdegree'].values.reshape(-1, 1))

#scaled_df['in_btc'] = StandardScaler().fit_transform(df['in_btc'].values.reshape(-1, 1))
#scaled_df['out_btc'] = StandardScaler().fit_transform(df['out_btc'].values.reshape(-1, 1))

rb_scaled_df['total_btc'] = preprocessing.RobustScaler(quantile_range=(25.0, 75.0), with_centering=True,
       with_scaling=True).fit_transform(rb_scaled_df['total_btc'].values.reshape(-1, 1))

rb_scaled_df['mean_in_btc'] = preprocessing.RobustScaler(quantile_range=(25.0, 75.0), with_centering=True,
       with_scaling=True).fit_transform(rb_scaled_df['mean_in_btc'].values.reshape(-1, 1))

rb_scaled_df['mean_out_btc'] = preprocessing.RobustScaler(quantile_range=(25.0, 75.0), with_centering=True,
       with_scaling=True).fit_transform(rb_scaled_df['mean_out_btc'].values.reshape(-1, 1))

rb_scaled_df.describe()
```

```python
pw_scaled_df = scaled_df.copy()
pw_scaled_df['indegree'] = preprocessing.PowerTransformer().fit_transform(pw_scaled_df['indegree'].values.reshape(-1, 1))
pw_scaled_df['outdegree'] = preprocessing.PowerTransformer().fit_transform(pw_scaled_df['outdegree'].values.reshape(-1, 1))
#scaled_df['in_btc'] = StandardScaler().fit_transform(df['in_btc'].values.reshape(-1, 1))
#scaled_df['out_btc'] = StandardScaler().fit_transform(df['out_btc'].values.reshape(-1, 1))
pw_scaled_df['total_btc'] = preprocessing.PowerTransformer().fit_transform(pw_scaled_df['total_btc'].values.reshape(-1, 1))
pw_scaled_df['mean_in_btc'] = preprocessing.PowerTransformer().fit_transform(pw_scaled_df['mean_in_btc'].values.reshape(-1, 1))
pw_scaled_df['mean_out_btc'] = preprocessing.PowerTransformer().fit_transform(pw_scaled_df['mean_out_btc'].values.reshape(-1, 1))
pw_scaled_df.describe()
```

```python
qt_scaled_df = scaled_df.copy()
qt_scaled_df['indegree'] = preprocessing.QuantileTransformer(output_distribution='normal').fit_transform(qt_scaled_df['indegree'].values.reshape(-1, 1))
qt_scaled_df['outdegree'] = preprocessing.QuantileTransformer(output_distribution='normal').fit_transform(qt_scaled_df['outdegree'].values.reshape(-1, 1))
#scaled_df['in_btc'] = StandardScaler().fit_transform(df['in_btc'].values.reshape(-1, 1))
#scaled_df['out_btc'] = StandardScaler().fit_transform(df['out_btc'].values.reshape(-1, 1))
qt_scaled_df['total_btc'] = preprocessing.QuantileTransformer(output_distribution='normal').fit_transform(qt_scaled_df['total_btc'].values.reshape(-1, 1))
qt_scaled_df['mean_in_btc'] = preprocessing.QuantileTransformer(output_distribution='normal').fit_transform(qt_scaled_df['mean_in_btc'].values.reshape(-1, 1))
qt_scaled_df['mean_out_btc'] = preprocessing.QuantileTransformer(output_distribution='normal').fit_transform(qt_scaled_df['mean_out_btc'].values.reshape(-1, 1))
qt_scaled_df.describe()
```
